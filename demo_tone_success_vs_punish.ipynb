```python
# demo_tone_success_vs_punish: imports

from core import (
    DEFAULT_DATA_ROOT,
    load_tone_success_punish_all,
    run_cebra,
    plot_cebra_3d,
)
```

```python
# demo_tone_success_vs_punish: load data + run CEBRA

root = DEFAULT_DATA_ROOT
print(f"Using data root: {root}")

(
    neural_tone_all_success,
    neural_tone_all_punish,
    tone_mat_paths_dcz,
    tone_mat_paths_saline,
) = load_tone_success_punish_all(root)

print("Success matrix shape:",
      None if neural_tone_all_success is None else neural_tone_all_success.shape)
print("Punish matrix shape:",
      None if neural_tone_all_punish is None else neural_tone_all_punish.shape)

if neural_tone_all_success is None or neural_tone_all_punish is None:
    raise RuntimeError("No success or punish trials found.")
```

```python
# demo_tone_success_vs_punish: CEBRA-Time + interactive plot

emb_success, emb_punish, model = run_cebra(
    model_type="time",
    dcz_data=neural_tone_all_success,
    saline_data=neural_tone_all_punish,
    model_architecture="offset10-model",
    temperature=1.0,
    output_dimension=3,
    max_iterations=2000,
)

plot_cebra_3d(
    emb_success,
    emb_punish,
    title="Tone Success vs Punish (CEBRA-Time)"
)
```
